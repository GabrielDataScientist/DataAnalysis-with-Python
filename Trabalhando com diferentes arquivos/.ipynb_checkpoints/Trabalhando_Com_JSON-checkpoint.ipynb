# Selecionando os cursos

```python
todas_matriculas = []
x = np.random.rand(20)
prob = x / sum(x)
```

```python
#Criando um loop para percorrer a qnt de matrículas por aluno (id) e escolher um curso aleatório no df cursos

#Primeiro loop percorre o df para coletar o id e qnt de matrículas
for index, linha in dados_nome.iterrows():
    id_aluno =  linha.ID_aluno
    matriculas = linha.matricula
    #Segundo loop percorre a qnt de matrículas alocando os cursos 
    for i in range(matriculas):
        mat = [id_aluno, np.random.choice(cursos.index, p = prob)]
        todas_matriculas.append(mat)
        
        
#Aloca o resultado em um novo dataframe
df_matricula = pd.DataFrame(todas_matriculas, columns = ['id_aluno', 'id_curso'])
```

```python
df_matricula
```

## Agrupando dois dataframes usando o groupby

Utiliza-se o join() para juntar os dataframes

```python
df_matricula.groupby('id_curso').count().join(cursos['Nome_curso']).rename(columns={'id_aluno':'Qnt._alunos_inscritos'})
```

```python
matriculas_por_curso = df_matricula.groupby('id_curso').count().join(cursos['Nome_curso']).rename(columns={'id_aluno':'Qnt._alunos_inscritos'})
```
