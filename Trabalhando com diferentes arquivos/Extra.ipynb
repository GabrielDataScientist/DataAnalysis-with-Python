```python
import pandas as pd
```

```python
df = pd.read_csv("nome_alunos.csv", sep=";")
```

```python
df
```

# Iterando sobre o dataframe

```python
for i in df:
    print(i)    
```

```python
#Retorna todos os dados da linha
for i, row in df.iterrows():
    print(f"Linha: {i}")
    print("")
    print(row)
    print("")
```

```python
#Retorna todos os dados de uma coluna especÃ­fica
for i in df['nome']:
    print(i)
```

```python
#Retornando dados de mais de uma coluna especificada
for linha, idaluno in df[['nome', 'ID_aluno']].iterrows():
    print(idaluno)
```

```python
#Retornando dados de mais de uma coluna especificada
for i, id_aluno in df[['nome', 'ID_aluno']].iterrows():
    print(f"Nome: {id_aluno['nome']} -> Id_aluno: {id_aluno['ID_aluno']}")
```
