# Incializando as bibliotecas necessárias

```python
import pandas as pd
import numpy as np
```

## JSON

```python
dados_fem = pd.read_json("https://guilhermeonrails.github.io/nomes_ibge/nomes-f.json")
dados_masc = pd.read_json("https://guilhermeonrails.github.io/nomes_ibge/nomes-m.json")
```

```python
dados_masc
```

```python
#Sorteando 5 dados aleatórios
dados_fem.sample(5)
```

```python
dados_masc.sample(5)
```

```python
#Contando a quantidade de nomes dentro dos daframes
print("Quantidade de nomes femininos: ", dados_fem['nome'].count())
print("Quantidade de nomes masculinos: ", dados_masc['nome'].count())
```

```python
#Outra forma de fazer a soma dos dois df juntos
print("Quantidade de nomes: " + str(len(dados_fem) + len(dados_masc)))
```

## Concatenando os dataframes

```python
frames = [dados_masc, dados_fem]
frames
```

```python
#Concatenando apenas os nomes
dados_nome = pd.concat(frames)['nome'].to_frame() #to_frame converte a variável para dataframe
dados_nome
```

```python
#Outra forma de concatenar os nomes
dados_nome = pd.concat([dados_masc, dados_fem])['nome'].to_frame()
dados_nome
```

## Gerando números aleatórios para o ID do aluno

```python
#Gera números pseudo-aleatórios
np.random.seed(123)
```

```python
#Tamanho do df
total_nomes = len(dados_nome)
total_nomes
```

```python
#Distribui de forma aleatória os números gerados, alocando na coluna ID
dados_nome['ID'] = np.random.permutation(total_nomes) + 1
```

```python
dados_nome[['nome', 'ID']]
```
