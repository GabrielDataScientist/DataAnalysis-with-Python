# Análise de Séries temporais usando o pandas e seaborn

```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Importando a base de dados

```python
df_alucar = pd.read_csv("base_dados_time_series/alucar.csv", sep=",")
df_alucar
```

```python
df_alucar.dtypes
```

### Convertendo a coluna mes de object para datetime

```python
df_alucar['mes'] = pd.to_datetime(df_alucar['mes'])
df_alucar.dtypes
```

### Visualizando os dados graficamente

```python
#Adicionando uma nova paleta de cor para o gráfico
sns.set_palette("Accent")
sns.set_style("darkgrid")
```

```python
#Plotando o gráfico
ax = sns.lineplot(x='mes', y='vendas', data = df_alucar)

#Alterando o tamanho do gráfico
ax.figure.set_size_inches(12,6)

#Alterando o título
ax.set_title("Vendas de carros", fontsize = 18)

#Alterando o valor da label de x e y
ax.set_xlabel("Mês", fontsize = 14)
ax.set_ylabel("Vendas(R$)", fontsize = 14)

#Para remover avisos do python
ax = ax
```

## Calculando a diferença de venda entre os meses

Para tal, usamos o diff, que calcula a diferença entre os elementos do dataframe. Por padrão, o diff faz o cálculo entre o elemento atual e o da linha anterior.

```python
df_alucar['diferenca'] = df_alucar['vendas'].diff()
df_alucar
```

```python
sns.set_style("white")
dif = sns.lineplot(x='mes', y='diferenca', data = df_alucar)
dif.figure.set_size_inches(12,6)

dif.set_title("Aumento das Vendas", fontsize=16)
dif.set_xlabel("Mes", fontsize=16)
dif.set_ylabel("Diferença", fontsize=16)
```

### Fazendo o mesmo plot usando uma função


```python
def plotar_grafico(x, y, dataset, title, xlabel, ylabel):
    #Padronizando o tipo do gráfico
    sns.set_palette("Accent")
    sns.set_style("dark")
    
    #Definindo o gráfico
    grafico = sns.lineplot(x=x, y=y, data = dataset)
    
    #Padronizando o tamanho
    grafico.figure.set_size_inches(12, 6)
    grafico.set_title(title, fontsize = 16)
    grafico.set_xlabel(xlabel, fontsize = 16)
    grafico.set_ylabel(ylabel, fontsize = 16)
    
    grafico = grafico

plotar_grafico("mes", "diferenca", df_alucar, "Diferença valor de vendas", "Mês", "Diferença")
```

## Calculando a oscilação da diferença de valor entre os meses

```python
#df_alucar['oscilacao'] = df_alucar['diferenca'].diff()
df_alucar
```

```python
plotar_grafico("mes", "oscilacao", df_alucar, "Oscilação das Vendas", "Mês", "Oscilação")
```

## Plotando os gráficos juntos

```python
#Definir o tamanho dos gráficos
plt.figure(figsize=(16,12))

#Definindo a qnt de gráficos a serem agrupados
# --- plt.subplot(número_de_linhas, número_de_colunas, índice_posicao_grafico)
gx = plt.subplot(3, 1, 1)
plotar_grafico("mes", 'vendas', df_alucar, "Venda de Carros", "Mês", "Venda(R$)")

#Segundo gráfico
gx = plt.subplot(3, 1, 2)
plotar_grafico("mes", "diferenca", df_alucar, "Diferença valor de vendas", "Mês", "Diferença")

#Terceiro gráfico
gx = plt.subplot(3, 1, 3)
plotar_grafico("mes", "oscilacao", df_alucar, "Oscilação das Vendas", "Mês", "Oscilação")
```
