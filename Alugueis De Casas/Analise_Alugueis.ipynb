## Removendo valores nulos

```python
dados_residenciais = pd.read_csv("aluguel_residencial.csv", sep=';')
dados_residenciais
```

```python
#Identificando a quantidade de registros nulos pelo info
dados_residenciais.info()
```

```python
#Utilizando o isnull() para identificar valores nulos (True)
dados_residenciais.isnull()
```

```python
#Utilizando o notnull para identificar valores nulos (False)
dados_residenciais.notnull()
```

```python
#Identificando valores nulos da coluna Valor
dados_residenciais[dados_residenciais['Valor'].isnull()]
```

```python
#Deletando os valores nulos
A = dados_residenciais.shape[0]
dados_residenciais.dropna(subset = 'Valor', inplace = True)
B = dados_residenciais.shape[0]

#Quantidade de dados deletados
print("Quantidade de dados deletados: {}".format(A - B))
```

```python
#Selecionando os apartamentos com valor de condomínio nulo
selecao = (dados_residenciais['Tipo'] == 'Apartamento') & (dados_residenciais['Condominio'].isnull())
```

```python
#Deletando os dados nulos. O til converte True em False e vice versa
A = dados_residenciais.shape[0]
dados_residenciais = dados_residenciais[~selecao]
B = dados_residenciais.shape[0]

print("Quantidade de linhas deletadas: {}".format(A - B))
```

```python
print("Qnt. de dados nulos do condomínio:", {dados_residenciais[dados_residenciais['Condominio'].isnull()].shape[0]})
print("Qnt. de dados nulos do iptu:", {dados_residenciais[dados_residenciais['IPTU'].isnull()].shape[0]})

#Atribuindo zero aos dados faltantes que sobraram
dados_residenciais = dados_residenciais.fillna({'Condominio': 0, 'IPTU': 0})

print("Qnt. de dados nulos pós tratamento do condomínio:", {dados_residenciais[dados_residenciais['Condominio'].isnull()].shape[0]})
print("Qnt. de dados nulos pós' tratamento do condomínio:", {dados_residenciais[dados_residenciais['IPTU'].isnull()].shape[0]})
```

```python
#Confirmando os dados
dados_residenciais.info()
```

# Salvando o arquivo com os dados tratados

```python
dados_residenciais.to_csv("aluguel_residencial.csv", sep = ';', index = False)
```

# CRIANDO NOVAS VARIÁVEIS

```python
dados_residenciais = pd.read_csv("aluguel_residencial.csv", sep = ';')
```

```python
#Valor bruto pago por mês
dados_residenciais['Valor Bruto'] = dados_residenciais['Valor'] + dados_residenciais['Condominio'] + dados_residenciais['IPTU']
dados_residenciais
```

```python
#Valor do metro quadrado
dados_residenciais['Valor m2'] = (dados_residenciais['Valor'] / dados_residenciais['Area']).round(2)
dados_residenciais
```

```python
#Valor do metro quadrado bruto
dados_residenciais['Valor m2 bruto'] = (dados_residenciais['Valor Bruto'] / dados_residenciais['Area']).round(2)
dados_residenciais
```

```python
#Alocando um valor para cada tipo de imóvel usando a função apply() e a função lambda
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
dados_residenciais['Tipo Agregado'] = dados_residenciais['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
dados_residenciais.rename(columns = {'Tipo Agregado':'Tipo Agregado'}, inplace = True)
```

Todos os valores que estiverem abaixo do primeiro quartil ou acima do quarto quartil são considerados discrepantes.

```python
valor = dados_residenciais['Valor']
```

```python
#Calculando o primeiro quartil
Q1 = valor.quantile(.25)
Q1
```

```python
#Calculando o terceiro quartil
Q3 = valor.quantile(.75)
Q3
```

```python
#Calculando o segundo quartil
IIQ = Q3 - Q1
IIQ
```

```python
#Limite inferior e superior
limite_inferior = Q1 - 1.5 * IIQ
limite_superior = Q3 + 1.5 * IIQ
```

```python
limite_inferior
```

```python
limite_superior
```

```python
selecao = (valor >= limite_inferior) & (valor <= limite_superior)
dados_novos = dados_residenciais[selecao]
```

## Analisando por tipo de imóvel

```python
dados_residenciais.boxplot(['Valor'], by=['Tipo'])
```

```python
#Agrupando por tipo
grupo_tipo = dados_residenciais.groupby('Tipo')['Valor']
type(grupo_tipo)
```

```python
Q1 = grupo_tipo.quantile(.25)
Q1
```

```python
Q3 = grupo_tipo.quantile(.75)
Q3
```

```python
Q2 = Q3 - Q1 #Mediana
Q2
```

```python
limite_inferior = Q1 - 1.5 * Q2
limite_superior = Q3 + 1.5 * Q2
```

```python
limite_inferior
```

```python
limite_superior
```

## Percorrendo o agrupamento 

```python
for tipo in grupo_tipo.groups.keys():
    print(tipo)
```

```python
#Criando um dataframe vazio
dados_finais = pd.DataFrame()

#Selecionando os imóveis dentro dos limites inferiores e superiores (excluindo os outliers)
for tipo in grupo_tipo.groups.keys():
    tipo_imovel = dados_residenciais['Tipo'] == tipo
    limites = (dados_residenciais['Valor'] >= limite_inferior[tipo]) & (dados_residenciais['Valor'] <= limite_superior[tipo])
    selecao = tipo_imovel & limites
    dados_selecao = dados_residenciais[selecao]
    dados_finais = pd.concat([dados_finais, dados_selecao])
```

```python
dados_finais.boxplot(['Valor'], by=['Tipo'])
```

## Exportando o arquivo
